# PyCity Schools Analysis

- This analysis provides information into how the amount spent, the size, and the type of school may have an affect on students and their reading/math test scores.

- The higher the spending range, the lower the overall passing score seems to drop.
- Schools that have smaller student sizes (Small or Medium) seem to yield better test resulst across the board.
- Per school type, it is shown that Charter schools yield higher test scores than District schools (Overall: 90.4% vs. 53.6%).
  
---

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("schools_complete.csv")
student_data_to_load = Path("students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = school_data_complete["school_name"].nunique()
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete["student_name"].count()
student_count
```

```python
# Calculate the total budget
total_budget = school_data_complete["budget"].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count/ float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({"Total Schools": [school_count],
                                "Total Students": [student_count],
                                 "Total Budget": [total_budget],
                                 "Average Math Score": [average_math_score],
                                 "Average Reading Score": [average_reading_score],
                                 "Students Passed Math": [passing_math_count],
                                 "Students Passed Reading": [passing_reading_count],
                                 "Total Students Passed": [overall_passing_rate]})

# Formatting
district_summary["Total Schools"] = district_summary["Total Schools"].map("{:,}".format)
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:.1f}%".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:.1f}%".format)
district_summary["Students Passed Math"] = district_summary["Students Passed Math"].map("{:,.0f}".format)
district_summary["Students Passed Reading"] = district_summary["Students Passed Reading"].map("{:,.0f}".format)
district_summary["Total Students Passed"] = district_summary["Total Students Passed"].map("{:.1f}%".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select all of the school types
school_types = school_data_complete.groupby("school_name")["type"].first()
school_types

```

```python
# Calculate the total student count per school
per_school_counts = school_data_complete.groupby("school_name")["student_name"].count()
per_school_counts
```

```python
# Calculate the total school budget and per capita spending per school
per_school_budget = school_data_complete.groupby("school_name")["budget"].first()
per_school_budget

per_school_capita = per_school_budget/per_school_counts
per_school_capita
```

```python
# Calculate the average test scores per school
per_school_math = school_data_complete.groupby("school_name")["math_score"].mean()
per_school_math

per_school_reading = school_data_complete.groupby("school_name")["reading_score"].mean()
per_school_reading
```

```python
# Calculate the number of students per school with math scores of 70 or higher
students_passing_math = school_data_complete[(school_data_complete["math_score"] >= 70)]
school_students_passing_math = students_passing_math.groupby(["school_name"]).size()
school_students_passing_math
```

```python
# Calculate the number of students per school with reading scores of 70 or higher
students_passing_reading = school_data_complete[(school_data_complete["reading_score"] >= 70)]
school_students_passing_reading = students_passing_reading.groupby(["school_name"]).size()
school_students_passing_reading
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
school_students_passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_students_passing_math / per_school_counts * 100
per_school_passing_reading = school_students_passing_reading / per_school_counts * 100
overall_passing_rate = school_students_passing_math_and_reading / per_school_counts * 100
print(per_school_passing_math.head())
print(per_school_passing_reading.head())
print(overall_passing_rate.head())
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({"School Types": school_types, 
                                "Total Students Per School": per_school_counts,
                                 "Total Budget Per School": per_school_budget,
                                 "Capita Per School": per_school_capita,
                                 "Average Math Score Per School": per_school_math,
                                 "Average Reading Score Per School": per_school_reading,
                                 "Students Passed Math Per School": school_students_passing_math,
                                 "Students Passed Reading Per School": school_students_passing_reading,
                                 "Total Students Passed Per School": school_students_passing_math_and_reading,
                                 "Overall Passing Rate Per School": overall_passing_rate})

# Formatting
per_school_summary["Total Students Per School"] = per_school_summary["Total Students Per School"].map("{:.0F}".format)
per_school_summary["Total Budget Per School"] = per_school_summary["Total Budget Per School"].map("${:,.2f}".format)
per_school_summary["Capita Per School"] = per_school_summary["Capita Per School"].map("${:,.2f}".format)
per_school_summary["Average Math Score Per School"] = per_school_summary["Average Math Score Per School"].map("{:.1f}%".format)
per_school_summary["Average Reading Score Per School"] = per_school_summary["Average Reading Score Per School"].map("{:.1f}%".format)
per_school_summary["Students Passed Math Per School"] = per_school_summary["Students Passed Math Per School"].map("{:,.0f}".format)
per_school_summary["Students Passed Reading Per School"] = per_school_summary["Students Passed Reading Per School"].map("{:,.0f}".format)
per_school_summary["Total Students Passed Per School"] = per_school_summary["Total Students Passed Per School"].map("{:,.0f}".format)
per_school_summary["Overall Passing Rate Per School"] = per_school_summary["Overall Passing Rate Per School"].map("{:.1f}%".format)

# Display the DataFrame
per_school_summary
```

## Highest-Performing Schools (by % Overall Passing)

## Bottom Performing Schools (By % Overall Passing)

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grade_math_scores = ninth_graders.groupby('school_name')
ninth_grade_math_avg = ninth_grade_math_scores['math_score'].mean()
tenth_grade_math_scores = tenth_graders.groupby('school_name')
tenth_grade_math_avg = tenth_grade_math_scores['math_score'].mean()
eleventh_grade_math_scores = eleventh_graders.groupby('school_name')
eleventh_grade_math_avg = eleventh_grade_math_scores['math_score'].mean()
twelfth_grade_math_scores = twelfth_graders.groupby('school_name')
twelfth_grade_math_avg = twelfth_grade_math_scores['math_score'].mean()

# Combine each of the scores above into single DataFrame called `math_scores_by_grade`
math_scores_by_grade = pd.DataFrame({"9th": ninth_grade_math_avg,
                                    "10th": tenth_grade_math_avg,
                                    "11th": eleventh_grade_math_avg,
                                    "12th": twelfth_grade_math_avg})

# Minor data wrangling
math_scores_by_grade.index.name = None

# Display the DataFrame
math_scores_by_grade
```

## Reading Score by Grade 

```python
# Use the code provided to separate the data by grade
ninth_graders = school_data_complete[(school_data_complete["grade"] == "9th")]
tenth_graders = school_data_complete[(school_data_complete["grade"] == "10th")]
eleventh_graders = school_data_complete[(school_data_complete["grade"] == "11th")]
twelfth_graders = school_data_complete[(school_data_complete["grade"] == "12th")]

# Group by `school_name` and take the mean of the the `reading_score` column for each.
ninth_grade_reading_scores = ninth_graders.groupby('school_name')
ninth_grade_reading_avg = ninth_grade_reading_scores['reading_score'].mean()
tenth_grade_reading_scores = tenth_graders.groupby('school_name')
tenth_grade_reading_avg = tenth_grade_reading_scores['reading_score'].mean()
eleventh_grade_reading_scores = eleventh_graders.groupby('school_name')
eleventh_grade_reading_avg = eleventh_grade_reading_scores['reading_score'].mean()
twelfth_grade_reading_scores = twelfth_graders.groupby('school_name')
twelfth_grade_reading_avg = twelfth_grade_reading_scores['reading_score'].mean()

# Combine each of the scores above into single DataFrame called `reading_scores_by_grade`
reading_scores_by_grade = pd.DataFrame({"9th": ninth_grade_reading_avg,
                                    "10th": tenth_grade_reading_avg,
                                    "11th": eleventh_grade_reading_avg,
                                    "12th": twelfth_grade_reading_avg})

# Minor data wrangling
reading_scores_by_grade = reading_scores_by_grade[["9th", "10th", "11th", "12th"]]
reading_scores_by_grade.index.name = None

# Display the DataFrame
reading_scores_by_grade
```

## Scores by School Spending

```python
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()
```

```python
#Change Total Capita Per School to str
school_spending_df["Capita Per School"] = school_spending_df["Capita Per School"].astype(str)
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(per_school_capita, spending_bins, labels=labels, include_lowest=True)
school_spending_df
```

```python
#Change format
school_spending_df["Average Math Score Per School"] = school_spending_df["Average Math Score Per School"].str.rstrip("%")
school_spending_df["Average Math Score Per School"] = pd.to_numeric(school_spending_df["Average Math Score Per School"])
```

```python
school_spending_df["Average Reading Score Per School"] = school_spending_df["Average Reading Score Per School"].str.rstrip("%")
school_spending_df["Average Reading Score Per School"] = pd.to_numeric(school_spending_df["Average Reading Score Per School"], downcast="float")
```

```python
school_spending_df["Students Passed Math Per School"] = school_spending_df["Students Passed Math Per School"].apply(lambda x : x.replace(',',''))
school_spending_df["Students Passed Math Per School"] = pd.to_numeric(school_spending_df["Students Passed Math Per School"], downcast="float")
```

```python
school_spending_df["Students Passed Reading Per School"] = school_spending_df["Students Passed Reading Per School"].apply(lambda x : x.replace(',',''))
school_spending_df["Students Passed Reading Per School"] = pd.to_numeric(school_spending_df["Students Passed Reading Per School"], downcast="float")
```

```python
school_spending_df["Overall Passing Rate Per School"] = school_spending_df["Overall Passing Rate Per School"].str.rstrip("%")
school_spending_df["Overall Passing Rate Per School"] = pd.to_numeric(school_spending_df["Overall Passing Rate Per School"], downcast="float")
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Math Score Per School"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Average Reading Score Per School"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Students Passed Math Per School"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Students Passed Reading Per School"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["Overall Passing Rate Per School"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({"Average Math Score": spending_math_scores,
                                "Average Reading Score": spending_reading_scores,
                                "Passed Math Total": spending_passing_math,
                                "Passed Reading Total": spending_passing_reading,
                                "Overall Passing Rate": overall_passing_spending})

# Display results
spending_summary
```
